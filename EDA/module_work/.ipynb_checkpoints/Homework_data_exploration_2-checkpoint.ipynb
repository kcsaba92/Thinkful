Are there any differences between the lunch types with respect to their performances in exams? If there are, how do you explain this?
Yes, according to the t-tests there are clearly differences between the scores of tests and the type of lunch. Maybe reduced and free lunch kids come from a lower socio economic status which affects their test scores.

```python
# Test whether group differences are significant.
lunch_type = df["lunch"].unique()
grouped_df = df.groupby("lunch")
for var in ["math score", "reading score", "writing score"]:
    print("------------------------------------------------")
    print("Comparisons for variable: {}".format(var))
    print("------------------------------------------------")
    for i in range(0, len(lunch_type)):
        for j in range(i+1, len(lunch_type)):
            print("t-test between groups {0} and {1}:".format(lunch_type[i], lunch_type[j]))
            print(stats.ttest_ind(
                df[df["lunch"]==lunch_type[i]][var], 
                df[df["lunch"]==lunch_type[j]][var]
            ))
```

Does the test preparation course seem to have an effect on the exam performances?
yes it had an effect on the scores.

```python
# Test whether group differences are significant.
test_type = df["test preparation course"].unique()
grouped_df = df.groupby("test preparation course")
for var in ["math score", "reading score", "writing score"]:
    print("------------------------------------------------")
    print("Comparisons for variable: {}".format(var))
    print("------------------------------------------------")
    for i in range(0, len(test_type)):
        for j in range(i+1, len(test_type)):
            print("t-test between groups {0} and {1}:".format(test_type[i], test_type[j]))
            print(stats.ttest_ind(
                df[df["test preparation course"]==test_type[i]][var], 
                df[df["test preparation course"]==test_type[j]][var]
            ))
```
